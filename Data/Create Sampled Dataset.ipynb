# Creating a sampled copy of full YouTube trending dataset

```python
import pandas as pd
df = pd.read_csv('US_youtube_trending_data.csv')
df.dropna()
df.drop_duplicates(inplace=True)
df_sampled = df.sample(frac=0.1)
df_sampled.head(5)
```

```python
df_sampled.to_csv('US_youtube_trending_data_sampled.csv', index=False)
```
